# Handling missing values

# Role category

```python
df["title"] = df["title"].str.lower()
df["title"].head()
```

```python
df["title"].value_counts().head(20)
```

```python
def categorize_role(title):
    if "data analyst" in title:
        return "Data Analyst"
    elif "business analyst" in title:
        return "Business Analyst"
    elif "product" in title and "analyst" in title:
        return "Product Analyst"
    elif "data scientist" in title:
        return "Data Scientist"
    elif "machine learning" in title or "ml engineer" in title or "ai/ml" in title:
        return "ML Engineer"
    elif "data engineer" in title:
        return "Data Engineer"
    else:
        return "Other"

df["role_category"] = df["title"].apply(categorize_role)

df["role_category"].value_counts()
```

```python
skill_counts = {
    "Python": df["python_required"].sum(),
    "SQL": df["sql_required"].sum(),
    "Excel": df["excel_required"].sum(),
    "Power BI": df["powerbi_required"].sum(),
    "Tableau": df["tableau_required"].sum(),
    "Machine Learning": df["ml_required"].sum(),
    "AWS": df["aws_required"].sum(),
    "Communication": df["communication_required"].sum()
}

skill_df = pd.DataFrame(skill_counts.items(), columns=["Skill", "Count"])
skill_df.sort_values(by="Count", ascending=False)
```

# Final Conclusion

This project analyzed 987 cleaned LinkedIn job postings across major data-related roles to identify hiring trends and skill demand patterns.

### 1. SQL Emerges as the Most Foundational Skill
SQL appears in the highest number of postings overall and dominates across Data Analyst (82.6%) and Data Engineer (88.9%) roles. This indicates that database proficiency is a core requirement across analytical and engineering positions.

### 2. Python Drives Advanced Technical Roles
Python demand is strongest in Data Scientist (88.1%) and ML Engineer (83.9%) roles, confirming its importance in modeling, automation, and advanced analytics workflows.

### 3. Data Analyst Roles Are Visualization-Heavy
Data Analyst postings show strong demand for Tableau (49.3%) and Power BI (40.3%), along with Excel (60%). This highlights the reporting and dashboard-oriented nature of analyst roles.

### 4. Machine Learning Is Role-Specific
Machine Learning is highly concentrated in ML Engineer (76.7%) and Data Scientist (67.2%) roles but minimal in Data Analyst (14.2%) and Business Analyst (6.2%) roles. This indicates clear specialization differences within data careers.

### 5. Cloud Skills Are Critical for Engineering Roles
AWS demand is highest in Data Engineer roles (52%), showing the growing importance of cloud infrastructure in modern data pipelines.

### 6. Soft Skills Matter
Communication is mentioned in over 50% of postings, demonstrating that technical proficiency alone is insufficient without strong stakeholder interaction skills.

---

### Overall Insight

The data reveals clear skill segmentation across roles:

- **Data Analyst → SQL + Excel + BI tools**
- **Data Scientist → Python + Machine Learning + SQL**
- **Data Engineer → SQL + Python + AWS**
- **ML Engineer → Python + ML + Cloud**
- **Business Analyst → Excel + SQL + Reporting tools**

These patterns reflect the structured specialization within modern data careers and provide practical guidance for skill prioritization based on role targeting.
